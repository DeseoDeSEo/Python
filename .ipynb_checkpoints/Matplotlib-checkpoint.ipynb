    ( 산점도, 막대그래프, 원그래프, 히스토그램 )




```python
import matplotlib.pyplot as plt
```

```python
y = [ 2, 4, 6 ]
# x라는 변수 축 범위 지정하지 않으면 자동으로 지정됨.
# x와 y의 변수개수가 같지 않으면 오류가 난다.
# plot- 선 그래프(Line plot)
plt.plot(y)
plt.show()
```

```python
y = [ 2, 4, 6 ]
x= [1,2,3]
# plot- 선 그래프(Line plot)
plt.plot(x,y)
plt.show()
```

### line style 
##### ( 줄여서 ls라고도 함.)


```python
plt.plot(x,y,linestyle='--')
plt.show()
```

# Marker

```python
plt.plot(x,y,marker='o')
plt.show()
```

```python
x=[2,4,6,8]
y=[10,13,16,20]
y2=[10,20,15,12]

plt.plot(x,y,ls='-', marker='o')     # 그래프 그리기
plt.plot(x,y2,ls=':')
plt.show()                # 그래프 보여주기    
```

# line width : 선 두께
( 줄여서 lw라고 사용 가능)

```python
plt.plot(x,y, lw=3)
plt.show()
```

# line color: 선 색상 변경


```python
plt.plot(x,y, color='#ACFADF', marker='D')
plt.show()
```

# marker color
( markerfacecolor줄여서 mfc로 가능 )

```python
plt.plot(x,y, color='#ACFADF', marker='D', mfc='#7C73C0')
plt.show()
```

# 마커 라인 색상 markeredgecolor -> mec

```python
plt.plot(x,y, color='#ACFADF', marker='D', mfc='#7C73C0', mec='black')
plt.show()
```

```python
import numpy as np
x=np.arange(7)
y=[1,4,5,8,9,5,3]
plt.plot(x,y, ls='--',lw=5, marker='o', mec='green', mew=6, mfc ='red', markersize='15')
```

# 그래프 범위 지정

```python
plt.plot(x,y, ls='--',lw=5, marker='o', mec='green', mew=6, mfc ='red', markersize='15')
plt.xlim(-2,8)
plt.grid() # 배경 격자무늬
plt.show()
```

# 틱의 위치를 보기 위해, 격자무늬 표현
## tick- 그래프 축의 위치

```python
plt.plot(x,y, ls='--',lw=5, marker='o', mec='green', mew=6, mfc ='red', markersize='15')

plt.xticks([0,3,6])   # x 축 틱 설정
plt.yticks([1,2,3])   # y 축 틱 설정
plt.grid() # 배경 격자무늬
plt.show()
```
